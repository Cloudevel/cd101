# Diagnóstico y monitoreo del sistema.

El diagnóstico y monitoreo de un sistema *GNU/Linux* permite conocer el estado de los recursos (CPU, memoria, disco, red) y las sesiones de usuario activas. En este notebook se presentan herramientas complementarias a las vistas en los notebooks de procesos (11) y volúmenes (18).

## Identidad del sistema.

### El comando ```hostname```.

El comando `hostname` muestra o establece el nombre del host del sistema.

````bash
hostname
````

### El comando ```hostnamectl```.

`hostnamectl` es la herramienta de *systemd* para gestionar el nombre del host. Permite ver y establecer el nombre estático, transitorio y descriptivo del sistema.

````bash
hostnamectl
````

### La man page de ```hostnamectl```.

````bash
man hostnamectl
````

## Monitoreo interactivo de procesos con ```htop```.

`htop` es un visualizador interactivo de procesos, más completo que `top`. Muestra en tiempo real el uso de CPU por núcleo, memoria, swap y la lista de procesos con colores.

Para instalarlo en sistemas basados en *Debian*:

```bash
sudo apt install htop
```

````bash
htop --version
````

### La man page de ```htop```.

````bash
man htop
````

## Estadísticas de memoria virtual con ```vmstat```.

El comando `vmstat` (*virtual memory statistics*) reporta información sobre procesos, memoria, paginación, I/O de bloques, interrupciones y actividad de CPU.

Sintaxis básica:

```bash
vmstat [intervalo] [cuenta]
```

````bash
vmstat
````

````bash
vmstat 1 3
````

````bash
vmstat -s
````

### La man page de ```vmstat```.

````bash
man vmstat
````

## Estadísticas de I/O con ```iostat```.

El comando `iostat` (*I/O statistics*) muestra estadísticas de uso de CPU y de I/O de dispositivos de almacenamiento. Forma parte del paquete `sysstat`.

Para instalarlo:

```bash
sudo apt install sysstat
```

````bash
iostat
````

````bash
iostat -x 1 3
````

### La man page de ```iostat```.

````bash
man iostat
````

## Uso de espacio en disco con ```du```.

El comando `du` (*disk usage*) estima el espacio utilizado por archivos y directorios. A diferencia de `df`, que reporta el uso a nivel de sistema de archivos, `du` opera a nivel de directorio.

Opciones más usadas:

| Opción | Descripción |
|--------|-------------|
| `-h` | Muestra tamaños en formato legible (KB, MB, GB) |
| `-s` | Muestra solo el total del directorio |
| `-a` | Muestra el uso de cada archivo |
| `--max-depth=N` | Limita la profundidad de directorios mostrada |

````bash
du -sh /var/log
````

````bash
du -h --max-depth=1 /var
````

````bash
du -h --max-depth=1 /var | sort -h
````

### La man page de ```du```.

````bash
man du
````

## Archivos abiertos con ```lsof```.

El comando `lsof` (*list open files*) muestra los archivos abiertos por los procesos en ejecución. En *GNU/Linux*, todo es un archivo: sockets, pipes, dispositivos y archivos regulares.

Usos comunes:

````bash
lsof | head -20
````

````bash
lsof -u $USER | head -20
````

````bash
lsof -i :22
````

### La man page de ```lsof```.

````bash
man lsof
````

## Ejecución periódica con ```watch```.

El comando `watch` ejecuta un comando de forma repetida con un intervalo definido y muestra su salida en pantalla. Es útil para monitorear cambios en tiempo real.

Sintaxis:

```bash
watch [-n segundos] comando
```

El intervalo predeterminado es de 2 segundos.

````bash
watch --version
````

**Ejemplos de uso** (ejecutar en terminal):

```bash
# Monitorear uso de disco cada 5 segundos
watch -n 5 df -h

# Monitorear los 5 procesos que más CPU consumen
watch -n 2 'ps aux --sort=-%cpu | head -6'
```

### La man page de ```watch```.

````bash
man watch
````

## Diagnóstico de red.

### El comando ```ping```.

El comando `ping` envía paquetes *ICMP ECHO_REQUEST* a un host para verificar la conectividad de red y medir la latencia.

````bash
ping -c 4 localhost
````

### El comando ```traceroute```.

El comando `traceroute` muestra la ruta que siguen los paquetes desde el equipo local hasta un destino, listando cada salto (*hop*) intermedio.

Para instalarlo:

```bash
sudo apt install traceroute
```

````bash
traceroute -m 5 localhost
````

### El comando ```mtr```.

`mtr` (*Matt's traceroute*) combina las funcionalidades de `ping` y `traceroute` en una sola herramienta interactiva que actualiza los resultados en tiempo real.

Para instalarlo:

```bash
sudo apt install mtr
```

Para obtener un reporte no interactivo:

````bash
mtr --report --report-cycles 3 localhost
````

## Sesiones de usuario con ```loginctl```.

El comando `loginctl` es parte de *systemd* y permite gestionar e inspeccionar las sesiones de usuario activas en el sistema.

````bash
loginctl list-sessions
````

````bash
loginctl list-users
````

### La man page de ```loginctl```.

````bash
man loginctl
````
