## Temporizadores.

Un temporizador (scheduler) en la tradiciﾃｳn de UNIX y GNU/Linux es un daemon que ejecuta tareas preestablecidas en intervalos de tiempo definidos.

El uso de los temporizadores permite realizar diversas acciones repetitivas sin la necesidad de ser ejecutadas manualmente.

Aﾃｺn cuando existen varios temporizadores para UNIX y GNU/Linux, los mﾃ｡s comﾃｺn es ```cron```.

## El daemon  ```crond```.

Los temporizadores estﾃ｡n ligados a un daemon el cual se encarga de ejecutar los comandos indicados a la hora definida.

**Ejemplo:**

En el caso de es que esta notebook sea ejecutada desde la mﾃ｡quina virtual proporcionada por Cloudevel<sup>ﾂｮ</sup>, es posible conocer el estado del servicio mediante la siguiente instrucciﾃｳn.

````bash
sudo systemctl status cron --no-pager
````

### Los archivos *crontab*.

Las tablas de cron o "crontabs" son archivos de texto que contienen una serie de lﾃｭneas que decriben la ejecuciﾃｳn de un comando especﾃｭfico en un tiempo determinado.

Cada lﾃｭnea del archivo crontab se apegarse al siguiente formato:

```
* * * * * comando(s)
- - - - -
| | | | |
| | | | ----- Dﾃｭa de la semana (0 - 7) (Sunday=0 or 7)
| | | ------- Mes (1 - 12)
| | --------- Dﾃｭa del mes(1 - 31)
| ----------- Hora (0 - 23)
------------- Minutos (0 - 59)
```

Cuando se utiliza el asterisco * , se estﾃ｡ haciendo la indicaciﾃｳn de que el comando aplica a todas las opciones de tiempo correspondientes.

Los archivos *crontab* se localizan en el directorio */var/spool/cron/*

````bash
sudo ls /var/spool/cron
````

**Ejemplo:**

La siguiente lﾃｭnea:

```
30   3    *    *    1 yum update  
```

Indica que se ejecutarﾃ｡ el comando ```yum update``` cada lunes a las 3:30 horas.

### El comando ```crontab```.

No se recomienda editar los archivos *crontab* directamente. El comando ```crontab``` permite manipular un  archivo de este tipo para el usuario actual en caso de que cuente con los permisos para realizar estas operaciones.

```
crontab <opciones> <argumentos>
```
Las principales opciones de este comando son:

* ```-e``` permite editar un archivo crontab para el usuario actual. En caso de que no exista, se crearﾃ｡ un archivo temporal.
* ```-l``` enlista el contendo del archivo crontab del usuario.
* ```-r``` elimina el archivo crontab.

**Ejemplos:**

* La siguiente celda desplegarﾃ｡ el contenido del archivo *crontab* del usuario actual.

Si no existe un archivo para el usuario, este no desplegarﾃ｡ nada.

````bash
crontab -l
````

* Ejecute el comando ```crontab -e``` desde una terminal y aﾃｱada la siguiente lﾃｭnea: 

```25 16 * * * echo "Saludos" >> ~/saludos.txt```

Lo anterior harﾃ｡ que el texto *Saludos" sea aﾃｱadido al archivo *~/saludos.txt* a las 14:25 horas diariamente.

````bash
crontab -l
````

````bash
ls ~/saludos.txt
````

````bash
cat ~/saludos.txt
````

* La siguiente celda elimininarﾃ｡ el crontab del usuario actual. 

````bash
crontab -r
````

````bash
crontab -l
````

## Archivos y directorios relevantes de ```crond```.

El daemon ```crond``` cuenta con algunos archivos que pueden ser utilizados por *root*.

### Archivos de permisos.

Los suguientes archivos delimitan a los usuarios que pueden o no utilizar crear *crontabs*.

* */etc/cron.allow* contiene una lista de los usuarios autorizados.
* */etc/cron.deny* contiene una lista de los usuarios autorizados.

````bash
sudo ls /etc/cron.deny
````

````bash
sudo ls /etc/cron.deny
````

````bash
sudo ls /etc/cron.allow
````

### Archivos de administraciﾃｳn.

Estos archivos contienen una serie de comandos que pueden ser ejecutados en un momento especﾃｭfico.

* */etc/cron.hourly* se ejecutarﾃ｡n en el primer segundo del primer minuto de cada hora.
* */etc/cron.daily* se jejecutarﾃ｡n en el primer segundo de un dﾃｭa, diarimante.
* */etc/cron.monthly* se ejecutarﾃ｡n en el primer segundo del primer dﾃｭa del mes, mensualmente.

````bash
ls /etc/cron.*
````

### Bitﾃ｡coras de cron.

El archivo *var/log/cron* contiene la bitﾃ｡cora de todas las acciones del ```crond```.

````bash
sudo cat /var/log/cron 
````
